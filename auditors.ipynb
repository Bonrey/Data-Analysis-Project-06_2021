```python
import re
```

```python
# функция поиска слова

def find_word(text, position, increment = True):
    word = ''
    val = 1
    if not increment: # инкрементируем или декрементируем счетчик
        val = -1
    while True:
        while text[position].isalpha() or text[position] == '«' or text[position] == '»' or text[position] == '"' or text[position] == '-':
            if increment:
                word += text[position]
            else:
                word = text[position] + word
            position += val
        if word != '':
            break
        else:
            position += val
    return word, position

# ищем название аудитора и ИНН аудитора

for el in found_word:
    all_pos = [m.start() for m in re.finditer('аудитор', texts_df.loc[el, 'text'], re.IGNORECASE)]
    name = ''
    number = ''
    for pos in all_pos:
        if (texts_df.loc[el, 'text'].find('утвердить', pos - 110, pos) != -1 or texts_df.loc[el, 'text'].find('Утвердить', pos - 150, pos) != -1) and (texts_df.text[el][pos + 8] == ' ' or texts_df.text[el][pos + 9] == ' '):
            text_now = texts_df.text[el]
            iterator = pos # мы на слове аудитор
            iterator -= 1
            if find_word(text_now, iterator, False)[0] != '«Утвердить' and find_word(text_now, iterator, False)[0] != 'Утвердить' and find_word(text_now, iterator, False)[0] != 'утвердить': # если предыдущее слово не "утвердить"

                while True:
                    word, iterator = find_word(text_now, iterator, False)
                    if word[-1] == '»' or word[-1] == '"':
                        while word != 'Утвердить' and word != 'утвердить':
                            name = word + ' ' + name
                            iterator -= 1
                            word, iterator = find_word(text_now, iterator, False)
                        break
                    elif word == 'Утвердить' or word == 'утвердить' or word == '«Утвердить':
                        break
                    else:
                        iterator -= 1
            
            if name == '':
                
                iterator = pos # мы на слове аудитор
                while text_now[iterator] != ' ':
                    iterator += 1
                iterator += 1

                word, iterator = find_word(text_now, iterator) # если мы на слове "Общества", то пропускаем его, иначе ищем следующее слово
                if word == 'Общества' or word == 'общества': # пропускаем это слово
                    iterator += 1 # пропускаем пробел
                    word, iterator = find_word(text_now, iterator)

                while True:
                    if word[0].isupper(): # если нашли слово с большой буквы
                        name = word
                        iterator += 1
                        word, iterator = find_word(text_now, iterator)
                        while word[-1] != '»' and word[-1] != '"': # пока не найдем слово, оканчивающееся кавычками
                            name += ' ' + word
                            iterator += 1
                            word, iterator = find_word(text_now, iterator)
                        name += ' ' + word # нашли название
                        break
                    else:
                        iterator += 1
                        word, iterator = find_word(text_now, iterator)

            # если есть ИНН, то вычленяем
            if name != '':
                iterator += 1
                if text_now[iterator] == '(':
                    while True:
                        word = ''
                        while text_now[iterator].isalpha():
                            word += text_now[iterator]
                            iterator += 1
                        if word == 'ИНН':
                            while not text_now[iterator].isnumeric():
                                iterator += 1
                            while text_now[iterator].isnumeric():
                                number += text_now[iterator]
                                iterator += 1
                            break
                        else:
                            if text_now[iterator] == ')':
                                break
                            iterator += 1
                            if text_now[iterator] == ')':
                                break
                break
            

    print("Номер текста:", el)
    print("Название аудитора:", name)
    print("ИНН аудитора:", number)
    print()
    
    if name != '':
        result_names[el] = name
    if number != '':
        result_numbers[el] = number
```

```python
# в данных текстах слово "отчетность" не имеет отношения к аудитору

found_word_2.remove(137) # отчетность не в том контексте
found_word_2.remove(263) # отчетность не в том контексте
```

```python
# в некоторых текстах не избирали аудитора, но есть информация про тип отчетности
# также удаляю текста, в которых нет названия аудитора

found_word_2 = list(set(found_word_2).intersection(set(found_word)))
no_name = set(found_word).difference(set(result_names.keys()))
found_word_2 = list(set(found_word_2).difference(set(no_name)))
```

```python
# слову "отчетность" всегда предшествует её тип
# почему-то при сортировке возникали проблемы, но так как очередность не влияет на работу программы - оставил проход по неотсортированному массиву

for el in found_word_2:
    all_pos = [m.start() for m in re.finditer('отчетности', texts_df.loc[el, 'text'], re.IGNORECASE)]
    for pos in all_pos:
        print('Номер текста:', el)
        print('Тип отчетности:', find_word(texts_df.text[el], pos - 1, False)[0])
        result_type[el] = find_word(texts_df.text[el], pos - 1, False)[0]
        print()
        break
```
