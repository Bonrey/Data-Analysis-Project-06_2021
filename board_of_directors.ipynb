```python
import re
```

```python
for i in range(texts_df.shape[0]):
    texts_df.loc[i, 'text'] = texts_df.iloc[i].text.replace(u'\xa0', u' ').replace(u'\t', u' ')
    texts_df.loc[i, 'text'] = re.sub(' +', ' ', texts_df.iloc[i].text)
```

```python
found_variant_1.remove(143)
found_variant_1.remove(216)
found_variant_1.remove(263)
found_variant_1.remove(373)

found_variant_2.remove(191)
```

```python
for el in found_variant_1:
    all_pos = [m.start() for m in re.finditer('Совет директоров', texts_df.loc[el, 'text'], re.IGNORECASE)]
    res = []
    flag_we_were_here = False
    for pos in all_pos:
        if texts_df.loc[el, 'text'].find('избрать', pos - 60, pos) != -1 or texts_df.loc[el, 'text'].find('Избрать', pos - 60, pos) != -1:
            flag_we_were_here = True
            text_now = texts_df.text[el]
            iterator = pos
            while text_now[iterator] != '\n' and text_now[iterator] != ':':
                iterator += 1
            iterator += 1
            if text_now[iterator] == ' ':
                iterator += 1
            if text_now[iterator] == '\n':
                iterator += 1
            while True:
                new_line = ''
                while text_now[iterator] != '\n' and text_now[iterator] != ',':
                    if str.isalpha(text_now[iterator]) or text_now[iterator] == ' ':
                        new_line += text_now[iterator]
                    iterator += 1
                iterator += 1
                new_line = new_line.lstrip()
                if (re.match(r'[А-ЯЁ][а-яё]+\s+[А-ЯЁ][а-яё]+(?:\s+[А-ЯЁ][а-яё]+)?', new_line)) and len(new_line.split()) < 6:
                    res.append(new_line)
                else:
                    break
            if res != []:
                break
    if flag_we_were_here and len(res) == 0:
        print(el, 'но рядом есть слово избрать!')
    else:
        print(el, res)
    if len(res) == 0:
        result_dict[el] = res
```
